```python
url = 'https://api.devhub.virginia.edu/v1/courses'
```

```python
import json
import urllib
import requests
import pprint
import pandas as pd 
```

```python
data = requests.get(url)
final = json.loads(data.text)
columns = final['class_schedules']['columns']
records = final['class_schedules']['records']
```

```python
subject = []
catalog_number = []
class_section = []
professor = []
term = []
year = []
first_name = []
last_name = []

# creates a list of integers that give each entry a unique record number 

record_number = list(range(len(records)))

for record in records:
    # gives shortened subject title  
    subject.append(record[0])
    
    # gives course number 
    catalog_number.append(record[1])
    
    # gives setion number if multiple sections
    class_section.append(record[2])
    
    # formats professor name 
    name = record[6].replace("'", "").replace('Jr.', '').replace('II', '').replace('III','').replace('IV', "").replace(" I", "").split(",")
    if len(name) > 1:
        correclty_ordered_name = name[1] + " " + name[0].strip()
        first_name.append(name[1])
        last_name.append(name[0])
        correclty_ordered_name = correclty_ordered_name.split(' ')
        name_final = correclty_ordered_name[0] + " " + correclty_ordered_name[-1]
    else:
        name_final = ''
    professor.append(name_final)
    
    #gives fall/summer/spring 
    term.append(record[12].split()[1])
    
    # gives year of the specific class 
    year.append(record[12].split()[0])

for i in range(len(first_name)):
    print(first_name[i] + " " + last_name[i])
```

```python
unique_profs = sorted(list(set(professor)))
for i in range(len(unique_profs)):
    if len(unique_profs[i].split()) > 1:
        print("(" + str(10000 + i) + ", '" + unique_profs[i].split()[0] + "', '" + unique_profs[i].split()[-1] + "'),")
    
```

```python
# combines previous values to give one input for classCode 
finalCourseCode = []
letters = pd.Series(data=subject)
numbers = pd.Series(data=catalog_number)
sect = pd.Series(data=class_section)

for i in range(len(letters)):
    final = letters[i] + numbers[i]
    finalCourseCode.append(final)
    
# when this is printed, the output can be copied into an insert statement to populate the class table in sql 
final = []
for i in range(len(finalCourseCode)):
    if finalCourseCode[i][:2] == "CS":
        if len(professor[i].split()) > 1:
            index = unique_profs.index(professor[i].split()[0] +  " " + professor[i].split()[-1])
            proff = str(10000 + index)
        else:
            proff = "NULL"
        what_to_insert = "('" + finalCourseCode[i] + "', '" + term[i] +"', " + year[i] + ", " +  proff +"),"
        if what_to_insert not in final:
            final.append(what_to_insert)
for each in final:
    print(each)

```
