```python
import json
import requests
import pprint
import pandas as pd 

data = requests.get(url)
final = json.loads(data.text)
columns = final['class_schedules']['columns']
records = final['class_schedules']['records']
```

```python
subject = []
catalog_number = []
class_section = []
professor = []
term = []
year = []

# creates a list of integers that give each entry a unique record number 

record_number = list(range(len(records)))

for record in records:
    # gives shortened subject title  
    subject.append(record[0])
    
    # gives course number 
    catalog_number.append(record[1])
    
    # gives setion number if multiple sections
    class_section.append(record[2])
    
    # formats professor name 
    name = record[6].replace("'", "").split(",")
    if len(name) > 1:
        correclty_ordered_name = name[1] + name[0]
    else:
        correclty_ordered_name = ''
    professor.append(correclty_ordered_name)
    
    #gives fall/summer/spring 
    term.append(record[12].split()[1])
    
    # gives year of the specific class 
    year.append(record[12].split()[0])

   
# combines previous values to give one input for classCode 
finalCourseCode = []
letters = pd.Series(data=subject)
numbers = pd.Series(data=catalog_number)
sect = pd.Series(data=class_section)

for i in range(len(letters)):
    final = letters[i] + numbers[i] + class_section[i]
    finalCourseCode.append(final)
    
# when this is printed, the output can be copied into an insert statement to populate the class table in sql 
for i in range(len(finalCourseCode)):
    print("(" + str(record_number[i]) + ", '" + finalCourseCode[i] + "', '" + term[i] +"', " + year[i] + ", '" + professor[i] +"'),")
```
